```python
def Caesar(key, message):
    cryptogram = ""
    for c in message:
        if 96 < ord(c) < 123:
            c_ = ord(c)-97 # int value of the character c : a=0, b=1, ..., z=25
            c_ = (c_ + key)%26 # shift of key positions (Cesar encryption)
            cryptogram += chr(c_ + 97) # ASCII coding
        else :
            cryptogram += c
    return cryptogram
```

```python
import random

def Substitution(key, message):
    # key is a one dimension table of size 26. key[0] is the encryption of 'a', etc.
    cipher = ""
    for c in message:
        if 'a' <= c <= 'z':
            i = ord(c)-97
            cipher += key[i]
        else :
            cipher += c
    return cipher

def gen_key():
    key = [chr(i+97) for i in range(26)]
    random.shuffle(key)
    return key  
    
    
def inv_key(key):
    # compute the decipher key (resp. cipher key) from the cipher key (resp. decipher key)
    key_inv = [0]*26
    for i in range(26):
        j = ord(key[i])-97
        key_inv[j] = chr(i+97)
    return key_inv
```

```python
import random

def compresse(message):
    new_message = ""
    for c in message:
        if c in ['à']:
            new_message += 'a'
            continue
        if c in ['ç']:
            new_message += 'c'
            continue
        if c in ['è', 'é', 'ê', 'ë']:
            new_message += 'e'
            continue
        if c in ['ï', 'î']:
            new_message += 'i'
            continue
        if c in ['ö', 'ô']:
            new_message += 'o'
            continue
        if c in ['ü', 'û', 'ù']:
            new_message += 'o'
            continue
        if c in ['\'']:
            #new_message += ' '
            continue
        if 'A' <= c <= 'Z':
            new_message += chr(ord(c)+32)
            continue
        if 'a' <= c <= 'z' :
            new_message += c
    return new_message

def gen_key(t):
    key = ""
    for i in range(t):
        key += chr(random.randint(97,122))
    return key

def inv_key(key):
    key_inv = ""
    for c in key:
        if c == 'a':
            key_inv += 'a'
        else:
            key_inv += chr(220 - ord(c)) # chr(26 - (ord(c)-97) + 97)
    return key_inv

def Vigenere(key, message):
    cipher = ""
    t = len(key)
    i = 0
    for c in message:
        cipher += chr((ord(c) + ord(key[i]) - 2*97)%26 + 97)
        i = (i + 1)%t
    while i > 0 :
        cipher += chr(random.randint(97,122))
        i = (i + 1)%t
        
    return cipher
```

```python
def ind_coincidence(text):
    new_text = compresse(text)
    n = len(new_text)
    occurence = [0]*26
    for c in new_text :
        occurence[ord(c) - 97] += 1
    return sum([x*(x - 1)/(n*(n - 1)) for x in occurence])
```
